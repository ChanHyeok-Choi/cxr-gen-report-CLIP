```python
from ignite.metrics import Rouge

m = Rouge(variants=["L", 2], multiref="best")

### Sample 1
candidate = "the cat is there".split()
references = [
    "there is a cat on the mat".split()
]

m.update(([candidate], [references]))

print(m.compute())

### Efficient U Net + CLIP Loss sample 1
m = Rouge(variants=["L", 2], multiref="best")
candidate = ['borderline', 'cardiomediastinal', 'cardiomegaly', 'without', '.', '.', '.', '.', 'edema', 'bilateral', 'bilateral', 'with', 'without', 'effusion']
references = ['Apical lordotic frontal view. Considering differences in technical factors XXXX stable cardiomediastinal silhouette with mild cardiomegaly. No focal alveolar consolidation, no definite pleural effusion seen. Dense left lower lung nodule suggests a previous granulomatous process. No typical findings of pulmonary edema.'.split()]

m.update(([candidate], [references]))

print(m.compute())

### Efficient U Net sample 1
m = Rouge(variants=["L", 2], multiref="best")
candidate = ['and', 'of', 'of', 'of', 'of', '.', '.', 'of', '.', '.', 'of', '.', '.', '.']
references = ['Apical lordotic frontal view. Considering differences in technical factors XXXX stable cardiomediastinal silhouette with mild cardiomegaly. No focal alveolar consolidation, no definite pleural effusion seen. Dense left lower lung nodule suggests a previous granulomatous process. No typical findings of pulmonary edema.'.split()]

m.update(([candidate], [references]))

print(m.compute())

### ResNet + CLIP Loss sample 1
m = Rouge(variants=["L", 2], multiref="best")
candidate = ['silhouette', 'size', 'and', 'size', 'cardiomediastinal', 'normal', '.', '.', '.', 'No', 'No', '.', 'No', '.']
references = ['The heart is normal in size and contour. There is no mediastinal widening. No focal airspace disease. No large pleural effusion or pneumothorax. The XXXX are intact.']
m.update(([candidate], [references]))

print(m.compute())

### ResNet Sample 1
m = Rouge(variants=["L", 2], multiref="best")
candidate = ['The', 'and', 'are', 'are', 'are', '.', '.', 'are', '.', 'in', '.', '.', '.', 'are']
references = ['The heart is normal in size and contour. There is no mediastinal widening. No focal airspace disease. No large pleural effusion or pneumothorax. The XXXX are intact.']

m.update(([candidate], [references]))

print(m.compute())
```

```python
import evaluate

### Sample 1
meteor = evaluate.load('meteor')
predictions = ["It is a guide to action which ensures that the military always obeys the commands of the party"]
references = ["It is a guide to action that ensures that the military will forever heed Party commands"]
results = meteor.compute(predictions=predictions, references=references)
print(results)

### Sample 2
predictions = ["A dog is there"]
references = ["There is a dog sitting on the mat"]
results = meteor.compute(predictions=predictions, references=references)
print(results)

### Efficient U Net + CLIP Loss sample 1
predictions = ['borderline', 'cardiomediastinal', 'cardiomegaly', 'without', '.', '.', '.', '.', 'edema', 'bilateral', 'bilateral', 'with', 'without', 'effusion']
predictions = [' '.join(x for x in predictions)]
references = ['Apical lordotic frontal view. Considering differences in technical factors XXXX stable cardiomediastinal silhouette with mild cardiomegaly. No focal alveolar consolidation, no definite pleural effusion seen. Dense left lower lung nodule suggests a previous granulomatous process. No typical findings of pulmonary edema.'.split()]
results = meteor.compute(predictions=predictions, references=references)
print(results)


### Efficient U Net sample 1
predictions = ['and', 'of', 'of', 'of', 'of', '.', '.', 'of', '.', '.', 'of', '.', '.', '.']
predictions = [' '.join(x for x in predictions)]
references = ['Apical lordotic frontal view. Considering differences in technical factors XXXX stable cardiomediastinal silhouette with mild cardiomegaly. No focal alveolar consolidation, no definite pleural effusion seen. Dense left lower lung nodule suggests a previous granulomatous process. No typical findings of pulmonary edema.'.split()]
results = meteor.compute(predictions=predictions, references=references)
print(results)

### ResNet + CLIP Loss sample 1
predictions = ['silhouette', 'size', 'and', 'size', 'cardiomediastinal', 'normal', '.', '.', '.', 'No', 'No', '.', 'No', '.']
predictions = [' '.join(x for x in predictions)]
references = ['The heart is normal in size and contour. There is no mediastinal widening. No focal airspace disease. No large pleural effusion or pneumothorax. The XXXX are intact.']
print(results)

### ResNet sample 1
predictions = ['The', 'and', 'are', 'are', 'are', '.', '.', 'are', '.', 'in', '.', '.', '.', 'are']
predictions = [' '.join(x for x in predictions)]
references = ['The heart is normal in size and contour. There is no mediastinal widening. No focal airspace disease. No large pleural effusion or pneumothorax. The XXXX are intact.']
results = meteor.compute(predictions=predictions, references=references)
print(results)
```
